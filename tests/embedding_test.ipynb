```python
POLICY_COLL_NAME = "NLPProj"
CHROMA_PATH = "./chromadb_data"

import chromadb

client = chromadb.PersistentClient(
    path= CHROMA_PATH
)
import os

os.environ["CUDA_VISIBLE_DEVICES"]="-1"
from langchain.embeddings import HuggingFaceInstructEmbeddings
embeddings = HuggingFaceInstructEmbeddings(
    model_name = "BAAI/bge-large-en",
    query_instruction = "Represent the request for legal acts related questions for retrieving of supporting information",
    embed_instruction = "Represent the request for legal document information for retrieval:" 
)
```

```python
from langchain.schema import Document
from pydantic import Field

docs = []
with open('Data_sachit.txt', 'r') as f:
    lines = f.readlines()
    distlines = {}

    for line in lines:
        if not line in distlines:
            line = line.replace("'","")
            line = line.strip()
            if not ("=== PDF START ===" in line or  "===PDF ENDS===" in line) :
                distlines[line]=""

                metadata = {'source': 'myinformation'}
                doc = Document(page_content = line, metadata= metadata)
                docs.append(doc)
```

```python
for doc in docs:
    print(doc)
```

```python
from langchain.vectorstores import Chroma

Chroma.from_documents(docs, collection_name=POLICY_COLL_NAME, persist_directory= CHROMA_PATH, embedding= embeddings, client=client)
chroma = Chroma(POLICY_COLL_NAME, persist_directory= CHROMA_PATH, embedding_function= embeddings)
chroma.persist()
```

```python
docs = chroma.similarity_search("Lok Sabha", k = 5)
print(docs)
```
